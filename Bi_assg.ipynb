```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
dt['Attrition'] = le.fit_transform(dt['Attrition'])
dt['Gender'] = le.fit_transform(dt['Gender'])
dt['OverTime'] = le.fit_transform(dt['OverTime'])
```

```python
# Drop irrelevant columns
dt = dt.drop(['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'], axis=1)
```

```python
dt = pd.get_dummies(dt,
                    columns=['BusinessTravel', 'Department', 'EducationField',
                             'JobRole', 'MaritalStatus'],
                    drop_first=True)
```

```python
#cheking processed data
dt.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = dt.drop('Attrition', axis=1)  # independent variables
y = dt['Attrition']               # target variable
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,       # 20% test, 80% train
    random_state=42,     # ensures reproducibility
    stratify=y           # ensures same class ratio in both sets (useful for classification)
)
```

```python
#confirmation
print("Training set shape:", X_train.shape)
print("Testing set shape:", X_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
dt = DecisionTreeClassifier(criterion='entropy', max_depth=5, random_state=42)

```

```python
dt.fit(X_train, y_train)
```

```python
y_pred = dt.predict(X_test)
# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Confusion Matrix
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Detailed classification report
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn import tree
import matplotlib.pyplot as plt

plt.figure(figsize=(25,10))  # make it large enough to read
tree.plot_tree(
    dt,                                 # your trained DecisionTreeClassifier
    filled=True,                        # color nodes based on class
    feature_names=X.columns,            # column names for readability
    class_names=['Stay', 'Leave'],      # your target classes
    rounded=True,                       # rounded node edges
    fontsize=8                          # readable text
)
plt.title("Decision Tree Visualization")
plt.show()
```

```python
# Predict labels for the test dataset
y_pred = dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Accuracy score
print("Accuracy:", accuracy_score(y_test, y_pred))

# Confusion matrix
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Detailed report (precision, recall, F1-score)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
